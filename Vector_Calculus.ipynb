```python
import numpy as np
from tabulate import tabulate
import matplotlib.pyplot as plt
```

${\Large \vec{r}(t)=2cos(t)\vec{i}-3sin(t)\vec{j}, 0≤t≤2π} \\ \color{navy}{x(t)=2cos(t) \\ y(t)=-3sin(t)} \\ \color{navy}{\vec{r}(t):\frac{x^2}{2^2}+\frac{y^2}{3^2}=1}, \color{maroon}{sin^2(t)+cos^2(t)=1}$

```python
t = np.arange(0,2*np.pi,0.01)
x = 2*np.cos(t)
y = -3*np.sin(t)
plt.plot(x,y, 'g')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of r(t)')
plt.show()
```

${\Large \vec{r}(t)=2cos(t)\vec{i}+2sin(t)\vec{j}+t\vec{k}, t≥0} \\ \color{navy}{x(t)=2cos(t) \\ y(t)=2sin(t) \\ z(t)=t} \\ \color{navy}{\vec{r}(t):x^2+y^2=2^2, z=t}, \color{maroon}{sin^2(t)+cos^2(t)=1}$

```python
fig = plt.figure()
ax = plt.axes(projection ='3d')
t = np.linspace(0, 10, 100)
x = 2*np.cos(t)
y = 2*np.sin(t)
z = t
ax.plot3D(x, y, z, 'g')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Graph of r(t)')
plt.show()
```
